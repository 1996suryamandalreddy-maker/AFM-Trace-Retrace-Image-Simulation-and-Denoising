```python
PI_FEATURES = meta['feature_columns_pi']
DD_FEATURE_COLS = meta['feature_columns_dd']
PI_BOUNDS = meta['pi_bounds']
TRIM = int(meta['center_trim'])
N_PIX = int(meta['n_pixels'])

FD_TABLE_CACHE = {}
def _fd_table_for(drive_nm):
    k = round(float(drive_nm), 4)
    if k not in FD_TABLE_CACHE:
        FD_TABLE_CACHE[k] = scanner.prepare_fd_table_for_drive(
            float(drive_nm), n_d=scanner_cfg.fd_n_d)
    return FD_TABLE_CACHE[k]

def predict_PI(drive_nm, setpoint, scan_speed_hat, I_gain_exp):
    """Stage-1 ridge prediction.  Returns clipped (P, I)."""
    f = {'log_scan_speed': float(np.log10(max(scan_speed_hat, 1e-12))),
         'drive_nm':       float(drive_nm),
         'setpoint':       float(setpoint),
         'log_I_exp':      float(np.log10(max(I_gain_exp, 1e-12)))}
    feats = np.array([[f[c] for c in PI_FEATURES]], float)
    log_PI = pi_ridge.predict(feats)[0]
    lp = float(np.clip(log_PI[0], PI_BOUNDS['bound_lo_P'], PI_BOUNDS['bound_hi_P']))
    li = float(np.clip(log_PI[1], PI_BOUNDS['bound_lo_I'], PI_BOUNDS['bound_hi_I']))
    return 10.0 ** lp, 10.0 ** li

def dd_feature_vec(drive_nm, setpoint, I_gain_exp):
    """Construct the 10-feature vector used by the data-driven and hybrid models."""
    d  = float(drive_nm); sp = float(setpoint); ig = float(I_gain_exp)
    A0 = float(scanner.get_A0_hat(d))
    log_ig = float(np.log10(max(ig, 1e-12)))
    return np.array([[d, sp, ig, log_ig, A0, np.log10(max(A0, 1e-12)),
                       d*sp, d*log_ig, sp*log_ig, A0*sp]], float)

def _center_line(y, trim=TRIM):
    y = np.asarray(y, float).copy()
    return y - np.nanmean(y[trim:-trim])

def predict_PI_scan_line(h_profile, drive_nm, setpoint,
                          scan_speed_hat=1.0, I_gain_exp=75.0):
    """DT scanner driven by the ridge-predicted (P, I).  Returns (trace, retrace, P, I)."""
    P, I = predict_PI(drive_nm, setpoint, scan_speed_hat, I_gain_exp)
    out = run_scanner_line(scanner, h_profile,
        drive_nm=float(drive_nm), setpoint=float(setpoint),
        scan_speed_hat=float(scan_speed_hat),
        P=float(P), I=float(I), cfg=scanner_cfg,
        fd_table=_fd_table_for(drive_nm))
    sim_tr, sim_rt = get_height_from_out(out)
    return sim_tr[:N_PIX], sim_rt[:N_PIX], P, I

def predict_data_driven(drive_nm, setpoint, I_gain_exp):
    """Pure data-driven prediction: features Ã¢â€ â€™ centered trace/retrace."""
    X     = dd_feature_vec(drive_nm, setpoint, I_gain_exp)
    Xpoly = dd_pack['scaler'].transform(dd_pack['poly'].transform(X))
    trace   = dd_pack['dd_trace'].predict(Xpoly).ravel()
    retrace = dd_pack['dd_retrace'].predict(Xpoly).ravel()
    return trace, retrace

def predict_hybrid(h_profile, drive_nm, setpoint,
                    scan_speed_hat=1.0, I_gain_exp=75.0):
    """PI + data-driven residual: DT prediction + learned correction."""
    sim_tr, sim_rt, P, I = predict_PI_scan_line(
        h_profile, drive_nm, setpoint, scan_speed_hat, I_gain_exp)
    X     = dd_feature_vec(drive_nm, setpoint, I_gain_exp)
    Xpoly = dd_pack['scaler'].transform(dd_pack['poly'].transform(X))
    res_tr = dd_pack['res_trace'].predict(Xpoly).ravel()
    res_rt = dd_pack['res_retrace'].predict(Xpoly).ravel()
    hyb_tr = _center_line(sim_tr) + res_tr
    hyb_rt = _center_line(sim_rt) + res_rt
    return hyb_tr, hyb_rt, P, I

print('predict_PI_scan_line, predict_data_driven, predict_hybrid ready')
```

**Step 2 Ã¢â‚¬â€ Generate synthetic AFM ground truth image**

```python
def generate_afm_ground_truth(
    size=256,
    pixel_size_nm=1.0,
    random_seed=42
):
    """
    Synthetic AFM-like ground truth height image.

    Includes:
    - flat areas
    - steps
    - sinusoidal height variations
    - roughness
    - dot/island patterns
    - line patterns
    - calibration grating-like structures

    Returns:
        image: 2D numpy array, height in nm
    """

    np.random.seed(random_seed)

    x = np.arange(size) * pixel_size_nm
    y = np.arange(size) * pixel_size_nm
    X, Y = np.meshgrid(x, y)

    image = np.zeros((size, size), dtype=np.float32)

    # -----------------------------
    # Flat base region
    # -----------------------------
    image += 5.0

    # -----------------------------
    # Step terraces
    # -----------------------------
    image[:, size//4:] += 15.0
    image[size//2:, :] += 10.0

    # -----------------------------
    # Sinusoidal height variation
    # -----------------------------
    sine_x = 8.0 * np.sin(2 * np.pi * X / 40.0)
    sine_y = 4.0 * np.sin(2 * np.pi * Y / 70.0)

    mask_sine = (X < size * 0.45) & (Y < size * 0.45)

    image[mask_sine] += sine_x[mask_sine] + sine_y[mask_sine]

    # -----------------------------
    # Smooth roughness
    # -----------------------------
    roughness = np.random.normal(
        loc=0.0,
        scale=2.0,
        size=(size, size)
    )

    roughness = gaussian_filter(roughness, sigma=1.2)

    image += roughness

    # -----------------------------
    # Dot / island patterns
    # -----------------------------
    dot_centers = [
        (180, 50),
        (210, 80),
        (170, 110),
        (220, 130),
        (190, 160)
    ]

    for cx, cy in dot_centers:
        radius = 12
        height = 20

        distance = np.sqrt((X - cx)**2 + (Y - cy)**2)

        dot = height * np.exp(
            -(distance**2) / (2 * (radius / 2)**2)
        )

        image += dot

    # -----------------------------
    # Horizontal line patterns
    # -----------------------------
    for i in range(20, 120, 20):
        image[150:155, i:i+60] += 18.0

    # -----------------------------
    # Vertical line patterns
    # -----------------------------
    for i in range(160, 240, 20):
        image[i:i+5, 20:120] += 12.0

    # -----------------------------
    # Calibration grating-like bars
    # -----------------------------
    grating_start_x = 150
    grating_end_x = 245
    grating_start_y = 180
    grating_end_y = 245

    period = 16
    bar_width = 8
    bar_height = 30

    for col in range(grating_start_x, grating_end_x, period):
        image[
            grating_start_y:grating_end_y,
            col:col+bar_width
        ] += bar_height

    # Make minimum height zero
    image = image - image.min()

    return image.astype(np.float32)
```

**Step 6 Ã¢â‚¬â€ Safe wrapper for RichardÃ¢â‚¬â„¢s scan-line simulator**

This function tries different calling styles automatically.

```python
def call_predict_PI_scan_line(h_profile, drive_nm, setpoint, scan_speed_hat=1.0, I_gain_exp=75.0):
    """
    Compatibility wrapper for Richard's predict_PI_scan_line function.

    This function now directly calls `predict_PI_scan_line` with the expected arguments.
    """
    try:
        return predict_PI_scan_line(h_profile, drive_nm, setpoint, scan_speed_hat, I_gain_exp)
    except TypeError as e:
        print("Could not call predict_PI_scan_line.")
        print("Function signature is:")
        print(inspect.signature(predict_PI_scan_line))
        print(f"Error: {e}")
        raise RuntimeError("Error calling predict_PI_scan_line. Check its signature and the arguments passed.") from e
```

**Step 8 Ã¢â‚¬â€ Simulate trace and retrace image row-by-row**

```python
def simulate_image_row_by_row(clean_image, drive_nm, setpoint, scan_speed_hat=1.0, I_gain_exp=75.0):
    """
    Apply Richard's 1D scan-line simulator to every row
    of a 2D AFM ground truth image.

    Input:
        clean_image: 2D image
        drive_nm: float, the drive_nm value for the scanner
        setpoint: float, the setpoint value for the scanner
        scan_speed_hat: float, optional, the scan speed (default 1.0)
        I_gain_exp: float, optional, the I gain exponent (default 75.0)

    Output:
        trace_image
        retrace_image
    """

    clean_image = np.asarray(clean_image, dtype=np.float32)

    n_rows, n_cols = clean_image.shape

    trace_image = np.zeros_like(clean_image, dtype=np.float32)
    retrace_image = np.zeros_like(clean_image, dtype=np.float32)

    for row in range(n_rows):

        h_line = clean_image[row, :]

        # Pass drive_nm and setpoint explicitly to call_predict_PI_scan_line
        output = call_predict_PI_scan_line(h_line, drive_nm, setpoint, scan_speed_hat, I_gain_exp)

        if isinstance(output, tuple) and len(output) >= 2:
            trace_line = output[0]
            retrace_line = output[1]
        else:
            raise ValueError(
                "predict_PI_scan_line should return trace and retrace lines."
            )

        trace_line = np.asarray(trace_line, dtype=np.float32)
        retrace_line = np.asarray(retrace_line, dtype=np.float32)

        # Safety check
        if trace_line.shape[0] != n_cols:
            trace_line = np.interp(
                np.linspace(0, len(trace_line)-1, n_cols),
                np.arange(len(trace_line)),
                trace_line
            )

        if retrace_line.shape[0] != n_cols:
            retrace_line = np.interp(
                np.linspace(0, len(retrace_line)-1, n_cols),
                np.arange(len(retrace_line)),
                retrace_line
            )

        trace_image[row, :] = trace_line
        retrace_image[row, :] = retrace_line

    return trace_image, retrace_image
```

**Step 12 Ã¢â‚¬â€ Define statistics: RMSE, MAE, SSIM**

```python
def compute_image_statistics(ground_truth, test_image, name="image"):
    """
    Compare test image to ground truth.

    Metrics:
    RMSE = Root Mean Squared Error
    MAE  = Mean Absolute Error
    SSIM = Structural Similarity Index
    """

    ground_truth = np.asarray(ground_truth, dtype=np.float32)
    test_image = np.asarray(test_image, dtype=np.float32)

    rmse = np.sqrt(np.mean((ground_truth - test_image) ** 2))
    mae = np.mean(np.abs(ground_truth - test_image))

    data_range = ground_truth.max() - ground_truth.min()

    ssim_value = ssim(
        ground_truth,
        test_image,
        data_range=data_range
    )

    print(f"Statistics for {name}")
    print("--------------------------------")
    print(f"RMSE: {rmse:.4f} nm")
    print(f"MAE : {mae:.4f} nm")
    print(f"SSIM: {ssim_value:.4f}")
    print()

    return {
        "name": name,
        "RMSE": rmse,
        "MAE": mae,
        "SSIM": ssim_value
    }
```

**Step 14 Ã¢â‚¬â€ Simple denoisers**

Here we try three common denoisers.

```python
def normalize_to_0_1(image):
    image = np.asarray(image, dtype=np.float32)

    min_val = image.min()
    max_val = image.max()

    if max_val - min_val == 0:
        return image * 0

    return (image - min_val) / (max_val - min_val)


def restore_original_range(denoised_norm, reference_image):
    ref_min = reference_image.min()
    ref_max = reference_image.max()

    return denoised_norm * (ref_max - ref_min) + ref_min


def denoise_gaussian(image, sigma=1.0):
    return gaussian_filter(image, sigma=sigma)


def denoise_median(image, size=3):
    return median_filter(image, size=size)


def denoise_tv(image, weight=0.1):
    image_norm = normalize_to_0_1(image)

    denoised_norm = denoise_tv_chambolle(
        image_norm,
        weight=weight
    )

    return restore_original_range(denoised_norm, image)


def denoise_wavelet_method(image):
    image_norm = normalize_to_0_1(image)

    denoised_norm = denoise_wavelet(
        image_norm,
        method="BayesShrink",
        mode="soft",
        rescale_sigma=True
    )

    return restore_original_range(denoised_norm, image)
```

# NEW SECTION: Simple synthetic AFM images

The original synthetic image mixes many structures in one scene. This section tests simpler images one at a time so trace/retrace differences are easier to see and interpret.


```python
def generate_simple_afm_ground_truth(kind="step", size=256, height_nm=20.0):
    """Generate simple AFM-like ground truth images in nm.

    Available kinds:
    - step: sharp height step
    - gaussian_bump: one smooth raised feature
    - two_bumps: two raised features with different width/height
    - sine_grating: periodic ridge pattern
    - trench: one narrow groove in a flat raised surface
    """
    x = np.linspace(-1, 1, size)
    y = np.linspace(-1, 1, size)
    X, Y = np.meshgrid(x, y)

    image = np.zeros((size, size), dtype=np.float32)

    if kind == "step":
        image[:, size // 2:] = height_nm

    elif kind == "gaussian_bump":
        sigma = 0.18
        image = height_nm * np.exp(-(X**2 + Y**2) / (2 * sigma**2))

    elif kind == "two_bumps":
        sigma1 = 0.12
        sigma2 = 0.20
        bump1 = height_nm * np.exp(-((X + 0.35)**2 + Y**2) / (2 * sigma1**2))
        bump2 = 0.6 * height_nm * np.exp(-((X - 0.35)**2 + Y**2) / (2 * sigma2**2))
        image = bump1 + bump2

    elif kind == "sine_grating":
        image = 0.5 * height_nm * (1 + np.sin(2 * np.pi * 8 * X))

    elif kind == "trench":
        image += height_nm
        trench_width = max(2, size // 24)
        center = size // 2
        image[:, center - trench_width:center + trench_width] = 0.0

    else:
        raise ValueError(f"Unknown simple image kind: {kind}")

    return image.astype(np.float32)
```

## Preview the simple ground truth images


```python
simple_image_kinds = ["step", "gaussian_bump", "sine_grating", "trench", "two_bumps"]

plt.figure(figsize=(16, 3.5))
for i, kind in enumerate(simple_image_kinds, start=1):
    img = generate_simple_afm_ground_truth(kind=kind, size=256)
    plt.subplot(1, len(simple_image_kinds), i)
    plt.imshow(img, cmap="viridis")
    plt.colorbar(label="Height nm", fraction=0.046, pad=0.04)
    plt.title(kind.replace("_", " ").title())
    plt.axis("off")

plt.tight_layout()
plt.show()
```

## Run the same trace/retrace and denoising workflow

This uses the same simulator, denoisers, and statistics functions defined earlier in the notebook.


```python
def run_simple_synthetic_case(kind, size=256, height_nm=20.0):
    ground_truth_simple = generate_simple_afm_ground_truth(
        kind=kind,
        size=size,
        height_nm=height_nm
    )

    trace_simple, retrace_simple = simulate_image_row_by_row(
        ground_truth_simple,
        drive_nm_value,
        setpoint_value
    )

    denoised_simple = {
        "trace_gaussian": denoise_gaussian(trace_simple, sigma=1.0),
        "trace_median": denoise_median(trace_simple, size=3),
        "trace_tv": denoise_tv(trace_simple, weight=0.08),
        "trace_wavelet": denoise_wavelet_method(trace_simple),
        "retrace_gaussian": denoise_gaussian(retrace_simple, sigma=1.0),
        "retrace_median": denoise_median(retrace_simple, size=3),
        "retrace_tv": denoise_tv(retrace_simple, weight=0.08),
        "retrace_wavelet": denoise_wavelet_method(retrace_simple),
    }

    case_stats = []
    images_to_score = {
        "trace_corrupted": trace_simple,
        "retrace_corrupted": retrace_simple,
        **denoised_simple
    }

    for image_name, image in images_to_score.items():
        stats_row = compute_image_statistics(
            ground_truth_simple,
            image,
            name=f"{kind}: {image_name}"
        )
        stats_row["case"] = kind
        stats_row["image"] = image_name
        case_stats.append(stats_row)

    return {
        "ground_truth": ground_truth_simple,
        "trace": trace_simple,
        "retrace": retrace_simple,
        "denoised": denoised_simple,
        "stats": case_stats,
    }
```

```python
simple_synthetic_results = {}
simple_synthetic_stats = []

for kind in simple_image_kinds:
    print("=" * 72)
    print(f"Running simple synthetic image: {kind}")
    result = run_simple_synthetic_case(kind)
    simple_synthetic_results[kind] = result
    simple_synthetic_stats.extend(result["stats"])

simple_synthetic_stats_df = pd.DataFrame(simple_synthetic_stats)
simple_synthetic_stats_df = simple_synthetic_stats_df[["case", "image", "RMSE", "MAE", "SSIM"]]
simple_synthetic_stats_df
```
